```
import numpy as np
import pandas as pd
import bar_chart_race as bcr
```

```
data= pd.read_csv("covid_19_india.csv") #reading csv file
data.head()                                                          
```

```
table = data.pivot_table(values='Confirmed',columns=['State/UnionTerritory'],
                         index='Date').replace(np.nan,0).astype(int)
table
```

```
table.sort_values(list(table.columns),inplace=True)
table=table.iloc[80:,]
table.drop(["Cases being reassigned to states"],inplace=True,axis=1)
table
```

```
# library for bar chart race is already called above
bcr.bar_chart_race(
    df=table,
    filename= None, #assign a name with .mp4 if you want to save the race as video format
    orientation='h',
    n_bars=10, sort='desc',
    fixed_order=False, fixed_max=True,
    steps_per_period=10,interpolate_period=False,
    label_bars=True,
    bar_size=.90,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    #period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total cases: {v.nlargest(39).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    #perpendicular_bar_func='median',
    period_length=500,
    figsize=(5,3 ),  dpi=144,
    cmap='gist_rainbow',
    title = "Covid-19 cases in India" , title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
    scale='linear',
    writer=None,fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=True)  
```
